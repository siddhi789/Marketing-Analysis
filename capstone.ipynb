### Q3) Analysis on attempts made to turn a person into successful depositor

```python
data['Response_in_Binary'] = np.where(data['y'] == 'no', 0 ,1)
cam = data[['Customer_id','Response_in_Binary','y','campaign']].groupby('campaign').agg({'Response_in_Binary': np.sum}).rename(columns={'Response_in_Binary': 'No_of_Customer_Converted'})
cam1 = data[['Customer_id','campaign']].groupby(['campaign']).count().rename(columns={'Customer_id' : 'Number_of_Customer_for_given_no_of_attempt'})
cam2 = pd.concat([cam1,cam], axis = 1)
cam2.reset_index(inplace=True)
cam2.rename(columns={'campaign' : 'No_of_attempt_made_on_a_customer'},inplace=True)
cam2["Percentage_conversion"] = round(cam2['No_of_Customer_Converted']/cam2['Number_of_Customer_for_given_no_of_attempt']*100,4)
```

```python
g1 = cam2[cam2['No_of_attempt_made_on_a_customer']<9]
g2 = cam2[cam2['No_of_attempt_made_on_a_customer'].isin(range(9,18))].reset_index(drop=True)
g3 = cam2[cam2['No_of_attempt_made_on_a_customer']>17].reset_index(drop=True)
```

```python
h1 = pd.DataFrame(g1[['Number_of_Customer_for_given_no_of_attempt','No_of_Customer_Converted']].agg({'Number_of_Customer_for_given_no_of_attempt':np.sum, 'No_of_Customer_Converted': np.sum})).transpose()
h1.insert(loc=0,column='Attempt Range', value='1 to 8')
h2 = pd.DataFrame(g2[['Number_of_Customer_for_given_no_of_attempt','No_of_Customer_Converted']].agg({'Number_of_Customer_for_given_no_of_attempt':np.sum, 'No_of_Customer_Converted': np.sum})).transpose()
h2.insert(loc=0,column='Attempt Range', value='9 to 17')
h3 = pd.DataFrame(g3[['Number_of_Customer_for_given_no_of_attempt','No_of_Customer_Converted']].agg({'Number_of_Customer_for_given_no_of_attempt':np.sum, 'No_of_Customer_Converted': np.sum})).transpose()
h3.insert(loc=0,column='Attempt Range', value='more than 17')
df_5 = pd.concat([h1,h2,h3], axis=0).reset_index(drop = True)
df_5["Percentage conversion to investor"] = round(df_5['No_of_Customer_Converted']/df_5['Number_of_Customer_for_given_no_of_attempt']*100,4)
df_5
```

##### There is no significant change in the no. of Investor if the large number of attempt is made on a single customer. It would be more effective if those large attempts are distributed among more number of customer than on a single customer.

```python
Total_no_of_attempts_made = df3['campaign'].sum()
Total_no_of_customer_converted_as_depositor = cam2['No_of_Customer_Converted'].sum()
d1 = {'Total_no_of_attempts_made': df3['campaign'].sum(),
     'Total_no_of_customer_converted_as_depositor' : data['Response_in_Binary'].sum()}
df6 = pd.DataFrame(d1,index=[0])
df6['Percentage_of_successful_attempt'] = round(df6['Total_no_of_customer_converted_as_depositor']/df6['Total_no_of_attempts_made']*100,4)
df6
```

###  Q5) Socio-economical analysis of the customers
#### (i) Consumer price index - monthly indicator 

```python
sc=pd.read_csv('Customer_social_economic_data_p1.csv')
sc
sc.info()
```

```python
sc1 = pd.merge(sc,data, on='Customer_id')
sc2 = sc1.groupby(['month']).agg({'cons.price.idx':np.mean})
sc2.reset_index(inplace=True)
dr1 = {'jan': 1,'feb': 2,'mar': 3,'apr': 4,'may': 5,'jun': 6,'jul': 7,'aug': 8,'sep':9,'oct': 10,'nov': 11,'dec': 12}
sc2['month in integer'] = sc2['month'].apply(lambda x:dr1[x])
sc2.sort_values('month in integer').reset_index(drop=True)
```

From the above DataFrame we can conclude that, Consumer price index increases from march to june hence the inflation increases, it goes down from june to august and have only slight fluctuation from august to november and finally in month of december decreases significantly which result in the more purchasing power of customer and hence the success of campaign.
Customers are dominated by group with consumer price index as 93.994


#### (ii) Consumer confidence index - monthly indicator 

```python
sc3 = sc1.groupby(['month']).agg({'cons.conf.idx':np.mean})
sc3.reset_index(inplace=True)
sc3['month in integer'] = sc3['month'].apply(lambda x:dr1[x])
sc3.sort_values('month in integer').reset_index(drop=True)
```

From the above DataFrame we can conclude that over the year Consumer confidence index increses means that customer believe that economy will grow and there job will be secure hence they will invest much which result in the success of campaign.

#### (iii) Euribor 3 month rate - daily indicator 

```python
sc4 = sc1.groupby(['month','day_of_week']).agg({'euribor3m':np.mean})
sc4.reset_index(inplace=True)
dr2 = {'mon':1,'tue':2,'wed':3,'thu':4,'fri':5,'sat':6,'sun':7}
sc4['month in integer'] = sc4['month'].apply(lambda x:dr1[x])
sc4['Week day in integer'] = sc4['day_of_week'].apply(lambda x:dr2[x])
sc4.sort_values(by=['month in integer','Week day in integer']).reset_index(drop = True)
```

#### From march till August Euribor rate increses means high interest rate for lending between the bank which is the indicator of economy slow down hence it will negatively affect the campaign. But it suddenly drop in september and then again increases till november finally it decreses in the end of year.

#### (iv) Number of employees - quarterly indicator 

```python
sc5 = sc1.groupby(['month']).agg({'nr.employed':np.mean})
sc5.reset_index(inplace=True)
sc5['month in integer'] = sc5['month'].apply(lambda x:dr1[x])
```

```python
quarter1 = sc5[sc5['month in integer']<4].agg({'nr.employed':np.mean})
quarter1['Quarter']='Quarter 1'
q1 = pd.DataFrame(quarter1).transpose()
quarter2 = sc5[sc5['month in integer'].isin(range(4,7))].agg({'nr.employed':np.mean})
quarter2['Quarter']='Quarter 2'
q2 = pd.DataFrame(quarter2).transpose()
quarter3 = sc5[sc5['month in integer'].isin(range(7,10))].agg({'nr.employed':np.mean})
quarter3['Quarter']='Quarter 3'
q3 = pd.DataFrame(quarter3).transpose()
quarter4 = sc5[sc5['month in integer']>9].agg({'nr.employed':np.mean})
quarter4['Quarter']='Quarter 4'
q4 = pd.DataFrame(quarter4).transpose()
sc6 = pd.concat([q1,q2,q3,q4], axis=0).reset_index(drop = True)
sc6
```

#### The mean number of employees remains constant (approximately) over the different quarter.

```python
## Subscription Rate by Age

lst = [data]
for column in lst:
    column.loc[column["age"] < 30,  'age_group'] = 20
    column.loc[(column["age"] >= 30) & (column["age"] <= 39), 'age_group'] = 30
    column.loc[(column["age"] >= 40) & (column["age"] <= 49), 'age_group'] = 40
    column.loc[(column["age"] >= 50) & (column["age"] <= 59), 'age_group'] = 50
    column.loc[column["age"] >= 60, 'age_group'] = 60
lst
```

```python
### Subscription Rate by Age 
count_age_response_pct = pd.crosstab(data['y'],data['age_group']).apply(lambda x: x/x.sum() * 100)
count_age_response_pct = count_age_response_pct.transpose() 


age = pd.DataFrame(data['age_group'].value_counts(),index=[30,40,50,20,60])

age['% Contacted'] = age['age_group']*100/age['age_group'].sum()
age['% Subscription'] = count_age_response_pct['yes']
age['% Not Subscription'] = count_age_response_pct['no']
age.drop('age_group',axis = 1,inplace = True)

age['age'] = [30,40,50,20,60]
age = age.sort_values('age',ascending = True)
age
```

#### Age Stage: Customer who are older are more likely to subscribe the term deposit because they need to prepare for their future life without income after retired.

##### People aged above 60 took up the greatest percentage in subscription, with around 40%. And young people under 20 also contributed a lot, while the middle aged only contributed less than half of the subscription. The main reason behind this situation is that old people always have a strong desire to save money after retirement, while young people have very limited disposable income.
